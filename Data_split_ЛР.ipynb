```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import random

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("beverage_r.csv", sep=";", index_col='numb.obs')
df.head()
```

```python
np.random.seed(seed=42)

r = np.random.uniform(low = 0.0, high=1.0, size=len(df))
```

**Задача 1. **
Разделить таблицу данных так, чтобы в обучающую выборку попало 66% наблюдений, а в тестовую выборку 33% наблюдений

```python
df_train = df.iloc[r >= np.quantile(r, 1/3), :]
df_test = df.iloc[np.logical_not(r >= np.quantile(r, 1/3)), :]

print(df_train, '\nКоличество элементов: ', len(df_train), ', ', round(len(df_train)/len(df), 3))
print(df_test, '\nКоличество элементов: ', len(df_test), ', ', round(len(df_test)/len(df), 3))
```

**Задача 2.**
Разделить таблицу данных на четыре одинаковых части.

```python
quantiles = np.quantile(r, [0.25, 0.5, 0.75])

part1 = df[r <= quantiles[0]]
part2 = df[(r > quantiles[0]) & (r <= quantiles[1])]
part3 = df[(r > quantiles[1]) & (r <= quantiles[2])]
part4 = df[r > quantiles[2]]

parts = [part1, part2, part3, part4]

for p in parts:
  print(p, '\nКоличество элементов: ', len(p), ', ', round(len(p)/len(df), 3))
```

**Задача 3.** Разделить таблицу данных на две стратифицированные подвыборки. По переменным кока-кола и минералка.

```python
df['stratify_col'] = df['COKE'].astype(str) + df['TAB'].astype(str)

df_train = pd.DataFrame()
df_test = pd.DataFrame()

for group, data in df.groupby('stratify_col'):
  r = np.random.uniform(low = 0.0, high=1.0, size=len(data))
  df_train = pd.concat([df_train, data.iloc[r >= np.quantile(r, 1/3), :]])
  df_test = pd.concat([df_test, data.iloc[np.logical_not(r >= np.quantile(r, 1/3)), :]])
```

```python
for group, data in df.groupby('stratify_col'):
  print('Для группы вида: ', group)
  print('  Тренировочная выборка: ', round(len(df_train[df_train['stratify_col'] == group])/len(df_train), 3))
  print('  Тестовая выборка: ', round(len(df_test[df_test['stratify_col'] == group])/len(df_test), 3))
  print('  Реальная выборка: ', round(len(df[df['stratify_col'] == group])/len(df), 3))
```
