2.7 Бросаются 4 игральные кости. Найти вероятность того, что на них выпадет по одинаковому числу очков.

$P(A) = \frac{6}{6^4} = 0.00463$

```python
random.seed(a=42, version=2)
n_exp = 100000

A = list(range(1,7))
k_success = 0

for i in range(n_exp):
  dice_counts = random.choices(A, k = 4)
  if len(set(dice_counts)) == 1:
    k_success +=1

print(f"Вероятность:  {round(k_success / n_exp, 5)}")

```

 2.8. В ящике 10 красных и 6 синих пуговиц. Какова вероятность
 того, что две наудачу вынутые пуговицы будут одноцветными?

$P(A) = \frac{10}{16}\cdot\frac{9}{15} + \frac{6}{16}\cdot\frac{5}{15} = \frac{1}{2}$

```python
random.seed(a=42, version=2)
n_exp = 100000

A = [1] * 10 + [2] * 6
k_success = 0

for i in range(n_exp):
  buttons = random.sample(A, 2)
  if buttons[0] == buttons[1]:
    k_success += 1

print(f"Вероятность:  {round(k_success / n_exp, 5)}")
```

 3.14. (Задача о встрече). Два лица A и B условились встретиться
 в определенном месте между двумя и тремя часами дня. Пришедший первым ждет другого в течение 10 минут, после чего уходит. Чему равна вероятность встречи этих лиц, если каждый из них может прийти в любое время в течение указанного часа?

$P(A) = \frac{60^2 - 50^2}{60^2} \approx 0.30556$

```python
random.seed(a=42, version=2)
n_exp = 100000
k_success = 0

for i in range(n_exp):
  times = np.random.uniform(0, 60, 2)
  if abs(times[0] - times[1]) <= 10:
    k_success +=1

print(f"Вероятность:  {round(k_success / n_exp, 5)}")
```

Задача 1.  Парадокс Монти Холла

 Представьте, что вы стали участником игры, в которой вам нужно выбрать
 одну из трёх дверей. За одной из дверей находится автомобиль, за двумя
 другими дверями—козы. Вы выбираете одну из дверей, например, номер 1,
 после этого ведущий, который знает, где находится автомобиль, а где— козы,
 открывает одну из оставшихся дверей, например, номер 3, за которой
 находится коза. После этого он спрашивает вас — не желаете ли вы изменить
 свой выбор и выбрать дверь номер 2?

Какая стратегия дает Вам наибольшие шансы выиграть автомобиль?

А) Принять предложение ведущего и изменить свой выбор.

Б) Отвергнуть предложение ведущего и сохранить свой выбор.

В) Все равно, что делать. Стратегии А) и Б) дают Вам одинаковые шансы.

Наибольшие шансы выиграть дает вариант А: поменять свой выбор.

Изначально мы выбираем дверь $d_1$ с вероятностью найти автомобиль $P_1 = \frac{1}{3}$. Шанс, что автомобиль за одной из двух оставшихся дверей $P_{23} = \frac{2}{3}$. После открытия третьей двери, шанс, что за дверью 2 (неоткрытой) будет автомобиль: $P2 = \frac{2}{3} ≈ 0.6667$.

```python
random.seed(a=42, version=2)
n_exp = 100000
k_success_change = 0
k_success_no_change = 0

A = list(range(1, 4))

for i in range(n_exp):
  avto_door = random.choice(A)
  chosen_door = random.choice(A)

  # nochange strategy
  if chosen_door == avto_door:
    k_success_no_change += 1

  # change strategy
  if chosen_door != avto_door:
    k_success_change += 1

print(f"Вероятность победы без смены двери:  {round(k_success_no_change / n_exp, 5)}")
print(f"Вероятность победы со сменой двери:  {round(k_success_change / n_exp, 5)}")
```

 Задача 2.
 Есть 1000 рублей. Срочно нужно еще 1000 рублей. Выход есть: играем в казино.

 Вероятность выигрыша равна 18/37, ничьих не бывает. Две стратегии

 а) один раз ставим 1000 рублей

 б) делаем ставки по одному рублю каждая, пока не выиграем 1000 рублей либо не проиграем все деньги.

 При какой стратегии вероятность выигрыша выше?

$P(A) = p = \frac{18}{37} \approx 0.48649$

$P(B) = p = \frac{1 - (\frac{q}{p})^{1000}}{1 - (\frac{q}{p})^{2000}} \approx 3.30297 \cdot 10^{-24} \approx 0$

```python
random.seed(a=42, version=2)
n_exp = 10000
k_success_1000 = 0
k_success_1 = 0

A = [1, -1]
A_weights = [18, 19]

for i in range(n_exp):
  # 1000 coins at once
  state = random.choices(A, weights=A_weights)
  if state == [1]:
    k_success_1000 +=1

  # 1 coin at once
  coin_sum = 1000

  check_flag = 100000
  while 0 < coin_sum < 2000 and check_flag > 0:
    state = random.choices(A, weights=A_weights)
    coin_sum += state[0]
    check_flag -= 1

  if coin_sum >= 2000:
    k_success_1 +=1


print(f"Вероятность победы, если ставить 1000 монет сразу:  {round(k_success_1000 / n_exp, 5)}")
print(f"Вероятность победы, если ставить по 1 монете за раз:  {round(k_success_1 / n_exp, 5)}")
```
